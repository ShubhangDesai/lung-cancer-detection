```python
luna_root = '/deep/group/shubhang/luna16/'
anns = pd.read_csv(luna_root + 'annotations.csv').as_matrix()
ids = np.unique(anns[:, 0])

len(ids)
```

```python
for cid, short_id in zip(ids, short_ids):
    if os.path.exists('/deep/group/shubhang/luna16/lungs/' + short_id + '.npy'):
        continue
    cands = anns[anns[:, 0] == cid]
    create_nodule_mask(luna_root + 'scans/' + cid + '.mhd', cands, short_id)
```

```python
from random import shuffle

ids = os.listdir(luna_root + 'lungs_batch')
shuffle(ids)
tot = len(ids)

train_split, dev_split = int(0.75 * tot), int(0.2 * tot)
train_ids, dev_ids, test_ids = ids[:train_split], ids[train_split:train_split+dev_split], ids[train_split+dev_split:]
datasets = {'train': train_ids, 'dev': dev_ids, 'test': test_ids}
```

```python
lung_src = luna_root + 'lungs_batch/'
gt_src = luna_root + 'gt_batch/'

for set_type in ['train', 'dev', 'test']:
    dataset = datasets[set_type]
    for cid in dataset:
        old_lung, old_gt = lung_src + cid, gt_src + cid
        new_lung, new_gt = lung_src + set_type + '/' + cid, gt_src + set_type + '/' + cid

        os.rename(old_lung, new_lung)
        os.rename(old_gt, new_gt)
```
